```

def get_hrtime():
    return (datetime.datetime.now() + datetime.timedelta(hours=2)).strftime('%Y-%m-%d_%H%M%S')
```

```
def dl_stocks(ticker:str):

  url = f"https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol={ticker}&interval=1min&apikey=VYMPBNXERFQNJ3CL&datatype=csv"
  response = requests.get(url)
  return response.content.decode("utf-8")
```

```

def download_data():
  for ticker in tickers:
    base_directory = f'/tmp/alpha/in/{ticker}/'
    timestamp = get_hrtime()
    file_prefix = ticker+'_'
    file_name = f"{file_prefix}{timestamp}.txt"
    file_path = base_directory + file_name
    file_contents = dl_stocks(ticker)
    print (file_path)
    dbutils.fs.put(file_path, file_contents, overwrite=True)
    time.sleep(1)
    print('_______')
    
download_data()
```

```
def stream_all(tickers):
  
  streams = {}
  
  for ticker in tickers:
    checkpoint_location = "/tmp/check_"+ticker

    stream = spark.readStream.text("/tmp/alpha/in/"+ticker)
    query = stream.writeStream.option("path", "/tmp/alpha/out/"+ticker) \
              .option("checkpointLocation", checkpoint_location) \
              .start()
    
    streams[ticker] = {
            "stream": stream,
            "query": query
        }
    
  return streams
```

```
streams = stream_all(tickers)
```

```
#fonction pour loader le dataframe et rajouter une colonne avec le nom de l'action

def create_dataset(action):
  df = spark.read.format("delta").load(f'dbfs:/tmp/alpha/out/{action}')
  df = df.withColumn("Action", lit(action))
  return df
```

```
#fonction pour spliter le dataset, supprimer la première colonne

def split_df(df):
  split_col = split(df["value"], ",")
#   df = df.withColumn("date", to_timestamp(split_col.getItem(0), "yyyy-MM-dd HH:mm:ss.SSS"))
  df = df.withColumn("date", split_col.getItem(0))
  df = df.withColumn("open", split_col.getItem(1).cast("float"))
  df = df.withColumn("high", split_col.getItem(2).cast("float"))
  df = df.withColumn("low", split_col.getItem(3).cast("float"))
  df = df.withColumn("close", split_col.getItem(4).cast("float"))
  df = df.withColumn("volume", split_col.getItem(5).cast("float"))
  df = df.drop("value")
  
  return df
```

```
#fonction pour supprimer la première ligne

def del_first_row(df):
  df = df.withColumn("row_id", monotonically_increasing_id())
  df = df.filter(df.row_id != 0)
  df = df.drop("row_id")
  return df
```

```
#fonction pour créer la "super-liste" de tous les dataframe

def create_super_list(dict_streams):
  dfs_list = []
  for action in dict_streams.keys():
    df = create_dataset(action)
    df = split_df(df)
    df = del_first_row(df)
    dfs_list.append(df)
  return dfs_list
```

```
#fonction pour créer un dataframe à partir de tous les dataframe de la liste

def create_final_df(dfs_list):
  dataAlphaVantage = dfs_list[0]  # Initialiser avec le premier DataFrame de la liste

  # ajouter un à un les dataframes restant 
  for df in dfs_list[1:]:
    dataAlphaVantage = dataAlphaVantage.union(df)
  
  return dataAlphaVantage
```

```
#On appelle les fonctions crées

liste_dataframes = create_super_list(streams)
alphavantage_df = create_final_df(liste_dataframes)
```
